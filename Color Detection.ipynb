```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read image
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Step 3: Resize for faster processing
img_small = cv2.resize(img, (150, 150))

# Step 4: Reshape and apply KMeans to find dominant color
pixels = img_small.reshape(-1, 3)
kmeans = KMeans(n_clusters=1, random_state=42)
kmeans.fit(pixels)
dominant_color = kmeans.cluster_centers_[0].astype(int)

# Step 5: Display image and its dominant color
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Uploaded Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(np.ones((100,100,3), dtype=np.uint8)*dominant_color)
plt.title(f"Dominant Color\nRGB = {tuple(dominant_color)}")
plt.axis('off')

plt.show()

```
